# list的長度

```python
my_list2 = ['pythan','javascript','C#']
print(len(my_list2)) #串列長度
print(len(my_list2[0])) #字串長度
```

# 附加

```python
my_list2 = ['pythan','javascript','C#']
my_list2.append('java')#加到最後一個
print(my_list2)
```

```python
my_list2 = ['pythan','javascript','C#']
my_list2.insert(0,'java')#加到指定位置
print(my_list2)
```

# 刪除

```python
my_list2 = ['pythan','javascript','C#']
my_list2.remove('pythan')#刪除指定元素
print(my_list2)

del my_list2[-1]#刪除指定位置的元素
print(my_list2)

my_list2.clear()#全部清除
print(my_list2)
```

## list取值

```python
a=[1,2,3,4,5,6,7,8,9]
b=a[0:3]#範圍取值 0,1,2 不包括3
print(b)
c=a[0:9:2] #間隔2取值
print(c)
d=a[0:]
print(d)
e=a[:9]
print(e)
f=a[:]#取全部值
print(f)
```

```python
a=[1,2,3,4,5,6,7,8,9]
del a[7:9]#刪除某範圍值 7,8
print(a)
print(min(a))#最小值
print(max(a))#最大值
print(a.index(1))#找出1所在的第一個索引值 a[0]
b=[1,0,0,1,1,1]
print(b.count(1))#出現的次數
a.reverse() #串列反轉
print(a)
a.sort()#串列由小到大排序,a的值會被改變 (串列被改變)
print(a)
```
