## for迴圈 :

```python
#1+2+3+4...+10總和
sum =0
for count in range(1,11,1): #從1開始 間隔1 不包含11
   sum+=count
print("1+2+3+4...+10=",sum)#空格會變成跟上面同一行
 
```

```python
#另一種寫法
#1+2+3+4...+10總和
sum =0
for count in range(11): #從1開始 間隔1 不包含11
   sum+=count
print("1+2+3+4...+10=",sum)
```

## 雞兔同籠 有35個頭 94隻腳 共幾隻兔跟雞

```python
for chicken in range(0,36): #範圍0-35
    rabbit=35-chicken  ##公式
    if(2*chicken+4*rabbit)==94:  #若所有腳等於94
        print("共有",chicken,"隻,兔共有",rabbit,"隻")
        break
else:
        print("無解")
        
```

## while迴圈

```python
sum=0
count=1
while count<=10:  ##定義count小於等於10運作
    sum+=count
    count+=1  ##每次加1
print("1+2+3+4+5..+10=",sum)
```

## 本金10000元 ,定存年利率0.6% 請問多久銀行存款會翻倍

```python
x=10000
years=0
while x<=20000:  ##小於等於20000
    x=x*1.006
    years+=1
print(years,"年以後","存款加倍 為:",int(x),"元")
```

## 帳號密碼判斷

```python
id="asd"  ##帳號
pwd="1234"  ##密碼
while True:  ##無窮迴圈
    x=input("帳號輸入:")  ##input輸入
    y=input("密碼輸入:")  
    if (x==id and y==pwd): #帳密正確時
        print("歡迎登入")
        break
    else:
        print("輸入錯誤")
```

## 亂數

```python
import random  ##亂數
for i in range(0,10):  ##範圍10
    print(random.randint(1,9))##亂數1-9
```

```python
#猜數字
import random
n=random.randint(1,10)  ##設亂數n
while True:
    guess=int(input("請輸入1-10"))##輸入數字
    if guess == n: #等於n
        print("你猜對了 數字是",n)
        break
    else:
        print("你猜錯了")    
```
