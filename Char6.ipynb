## 檔案處理
os套件有自己的檔案處理命令，而比較常用的是phthon內建的open
* w=可寫(預設建立檔案)，若檔案存在則清除舊檔案夾
* w+= 可寫可讀(預設建立檔案) 若檔案存在則清除舊檔案夾
* r=可讀
* r+=可寫可讀 但寫入資料指向開頭
* a=可寫不能讀 寫入資料指向尾端
* a+=可寫可讀 寫入資料指向尾端
* b=轉為二進位文件
* ab+ =2進位文件的讀寫 寫入資料指向尾端



```python
a=open("a.text","w")#新增檔案 open套件
a.write("this is a text\t")#write不會換行
a.write("this is a text2\n")
a.write("this is a text3\n")
a.close() #關閉檔案
```

## 檢查檔案是否存在

```python
os.path.isfile("a.text")
```

## 讀取檔案

```python
a=open("a.text","r")
print(a.readline())
print(a.readline())
print(a.readline())#多餘的命令 檔案只有2行
a.close() 
```

```python
a=open("a.txt","r")
print(a.readlines())#一次讀取多行 與a.reod()命令相同
a.close() 
```

```python
#加上檔案是否存在的判斷
if os.path.isfile("a.text"):
  a=open("a.text","r")
  print(a.readlines())
  a.close()
else:
  print("檔案不存在")  
```

## split函式

```python
if os.path.isfile("a.text"):
  a=open("a.text","r")
  for line in a:
    fields=line.split(" ")#以空白為切割欄位
    print(fields)
  a.close()
else:
  print("檔案不存在")  
```

## 附加檔案

```python
a=open("a.txt","a")
a.write("this is text4\n")
a.close() 
```

```python
a=open("a.txt","r")
print(a.readlines())
a.close() 
```

## 比較a與r+的不同
* a寫在檔案的結尾
* r+寫在檔案的開頭

```python
a=open("a.txt","r+")
a.write("this is text5\n")
a.close() 
```

```python
a=open("a.txt","r")
print(a.readlines())
a.close() 
```

## 刪除檔案

```python
if os.path.isfile("a.text"):
  os.remove("a.text")
  print("檔案刪除成功")
else:
  print("找不到檔案")
  print(os.path.isfile("a.text"))
```
