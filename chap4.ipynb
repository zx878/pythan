## 為何需要numpy?

 numpy由c撰寫,速度快python 10-1000倍以上
 numpy儲存結果較正確,python數字儲存不正卻
 numpy節省儲存空間
 numpy向量(矩陣計算)速度快
 numpy取值方便

```python
a_list=[1,2,3,4,5]
print(a_list*3)
```

```python
#numpy 套件的ndrray型態,提供element.-wlse的操作 每一元素
import numpy as np
a=np.array([1,2,3,4,5])
print(a*3)
```

```python
a=np.array([1,2,3,4,5])
print(type(a))
print(a.shape)#ndaary的形狀
print(a.dtype)#資料型態
print(a.ndim)#維度
```

```python
import numpy as np
z=np.zeros(6)#6個0
print(z) 
print(np.empty((2,4)))#2個二維陣列4筆資料
print(np.empty((2,3,4)))#2個三維陣列3陣列4筆資料
print(np.arange(10))#產生0-9
```

### ndarray的進階操作


```python
my_array=np.arange(10)
print(my_array)
print(my_array[0])#第一個
print(my_array[0:5])#第一到 5的前一個
```

### 二維


```python
my_2d_array=np.array([np.arange(0,5),np.arange(5,10)])#二維陣列0-4 5-9
print(my_2d_array)
print(my_2d_array[1,:])#第1的陣列全部
print(my_2d_array[:,1])#兩個陣列的第1個
print(my_2d_array[1,1])#第1陣列的第1個資料
```

```python
my_array=np.arange(0,10)
print(my_array)
print(my_array.shape)
my_2d_array=my_array.reshape(2,5)#重朔成2*5陣列
print(my_2d_array)
print((np.arange(0,100).reshape(10,10)))#0-99重朔成10*10陣列
```
